```python
import pandas as pd 
import seaborn as sns 
import matplotlib.pyplot as plt
```

```python
df=pd.read_excel(r"C:\Users\aparn\OneDrive\Desktop\adolution\Advolution 9.0 Dataset.xlsx",sheet_name='Death Timeline of the Patients')
unique_values_set = set(df['State'].unique())
print(unique_values_set)
```

```python
# Assuming 'Deat of the Patient Post Treatment' is the column you want to modify
df['Deat of the Patient Post Treatment'] = df['Deat of the Patient Post Treatment'].replace('-', 0)

# Convert the column to numeric, coercing errors to NaN
df['Deat of the Patient Post Treatment'] = pd.to_numeric(df['Deat of the Patient Post Treatment'], errors='coerce')

# Now, do the same for 'Total Patients Admitted'
df['Total Patients Admitted'] = df['Total Patients Admitted'].replace('-', 0)
df['Total Patients Admitted'] = pd.to_numeric(df['Total Patients Admitted'], errors='coerce')


# Now, you can proceed with your calculations
rest = []
for state in unique_values_set:
    state_data = df[df['State'] == state]
    tcp = state_data['Deat of the Patient Post Treatment'].sum()
    tt = state_data['Total Patients Admitted'].sum()
    
    # Calculate the ratio in percentage
    res = (tcp / tt) * 100
    
    rest.append({'State': state, 'percentage': res})

# 'rest' now contains the State-wise ratio in percentage
```

```python
reu=pd.DataFrame(rest)
print(reu)
```

```python
data = pd.DataFrame(rest)

# Set up the plot
plt.figure(figsize=(12, 6))
sns.barplot(x='State', y='percentage', data=data, palette='viridis')

# Customize the plot
plt.title('State-wise Percentage Ratios')
plt.xlabel('State')
plt.ylabel('Percentage Ratio')
plt.xticks(rotation=45, ha='right')

# Show the plot
plt.tight_layout()
plt.show()
```

```python
data.sort_values(by='percentage', ascending=False, inplace=True)
top_10 = data.head(10)
plt.figure(figsize=(12, 6))
sns.barplot(x='State', y='percentage', data=top_10, palette='viridis')

# Customize the plot
plt.title('State-wise Percentage Ratios')
plt.xlabel('State')
plt.ylabel('Percentage Ratio')
plt.xticks(rotation=45, ha='right')

# Show the plot
plt.tight_layout()
plt.show()
```

```python
data.sort_values(by='percentage', ascending=True, inplace=True)
top_10 = data.head(10)
plt.figure(figsize=(12, 6))
sns.barplot(x='State', y='percentage', data=top_10, palette='viridis')

# Customize the plot
plt.title('State-wise Percentage Ratios')
plt.xlabel('State')
plt.ylabel('Percentage Ratio')
plt.xticks(rotation=45, ha='right')

# Show the plot
plt.tight_layout()
plt.show()
```

```python
unique_year=set(df['Year'].unique())
print(unique_year)
```

```python
result=[]
for year in unique_year:
  state_year = df[df['Year'] == year]
  tttp = state_year['Deat of the Patient Post Treatment'].sum()
  tnt=  state_year['Total Patients Admitted'].sum()
  rat= (tttp/tnt)*100;
  result.append({'Year': year, 'percentage change': rat})
resultit=pd.DataFrame(result)
resultit.head()
```

```python
plt.figure(figsize=(10, 6))
plt.plot(resultit['Year'], resultit['percentage change'], marker='o', linestyle='-', color='b')

# Customize the plot
plt.title('Percentage Change in Deaths Post Treatment Over the Years')
plt.xlabel('Year')
plt.ylabel('Percentage Change (%)')
plt.grid(True)

# Show the plot
plt.show()
```
