```python
unique_values_set = set(df['State'].unique())
rest = []

for state in unique_values_set:
    state_data = df[df['State'] == state]
    state_data2 = df2[(df2['State'] == state) & (df2['Account of Crime'] == 'TVC- Rape')]
    state_data3 = df2[(df2['State'] == state) & (df2['Account of Crime'] == 'TVC- Riots')]
    state_data4 = df2[(df2['State'] == state) & (df2['Account of Crime'] == 'TVC- Robbery')]
    state_data5 = df2[(df2['State'] == state) & (df2['Account of Crime'] == 'TVC- Total')]

    mdh = state_data['Murder and Homicide'].sum()
    ro = state_data['Robbery'].sum()
    at = state_data['Auto Theft'].sum()
    dmd = state_data['Drug Manufacturing and Distribution'].sum()
    ro2 = state_data['Reckless Driving'].sum()
    tee = state_data['Tampering with evidence'].sum()
    co = state_data['Conspiracy'].sum()
    po = state_data['Possession of Illegal Weapon'].sum()
    tv = state_data['Tax Evasion'].sum()

    tr = state_data2['Account of Crime'].count()  # Change 'sum()' to 'count()' if you want to count occurrences
    trr = state_data3['Account of Crime'].count()
    tro = state_data4['Account of Crime'].count()
    tot = state_data5['Account of Crime'].count()

    rest.append({'State': state, 'Murder and Homicide': mdh, 'Robbery': ro, 'Auto Theft': at,
                 'Drug Manufacturing and Distribution': dmd, 'Reckless Driving': ro2,
                 'Tampering with evidence': tee, 'Conspiracy': co,
                 'Possession of Illegal Weapon': po, 'Tax Evasion': tv,
                 'TVC- Rape': tr, 'TVC- Riots': trr, 'TVC- Robbery': tro, 'TVC- Total': tot})
```
