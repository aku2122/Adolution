```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
df=pd.read_excel(r"C:\Users\aparn\OneDrive\Desktop\adolution\Advolution 9.0 Dataset.xlsx")

```

```python
df.head()
```

```python
unique_values_set = set(df['State'].unique())
tcf=0;
tt=0;
rest=[]
```

```python
for state in unique_values_set:
    state_data = df[df['State'] == state]
    tcp = state_data['Trial of Violent Crimes by Courts By Confession'].sum()
    tt = state_data['Trial of Violent Crimes by Courts Total'].sum()
    res = tcp / tt
    rest.append({'State': state, 'Ratio': res})

# Create a DataFrame after the loop
result_df = pd.DataFrame(rest)

# Print the result DataFrame
print(result_df)
```

```python
result_df.head()
```

```python
# Assuming 'result_df' is your DataFrame
result_df_sorted = result_df.sort_values(by='Ratio', ascending=False)

# Print the sorted DataFrame
print("DataFrame sorted by Ratio:")

print(result_df_sorted)
```

```python
result_df_sorted=result_df_sorted[0:28]
navg = result_df['Ratio'].mean()
plt.figure(figsize=(15, 6))

# Bar plot for the Ratio
plt.bar(result_df['State'], result_df['Ratio'], label='Ratio', color='blue')

# Line plot for the Average
plt.axhline(y=navg, color='red', linestyle='--', label='Average')

plt.title('Ratio and Average for Different States')
plt.xlabel('State',fontweight='bold')
plt.ylabel('Value',fontweight='bold')
plt.legend()
plt.xticks(rotation=45, ha='right')  # Rotate x-axis labels for be
for label in plt.xticks()[1]:
    label.set_fontweight('bold')
for label in plt.yticks()[1]:
    label.set_fontweight('bold')
```
